## 1.2.1. Typical Neural Network Architecture:
The typical neural network architecture consists of several layers ([Sakip S., 2019](https://www.preprints.org/manuscript/201811.0546)).

### 1.2.1. Input Layer
The input layer is the initial layer through which the neural network receives its input X, which represents the data from which it learns.

The number of neurons (nodes) in this layer is typically equal to the number of features in the input data, X.

Ideally, each input neuron represents one element in the vector x.

Basically, the input layer prepares the data for further processing by the subsequent layers.

### 1.2.2. Output Layer
This is the last layer of the neural network which outputs a vector y representing the result that the neural network came up with. The entries in this vector represent the values of the neurons in the output layer. Each neuron in the last layer would represent a different class.

For instance, for a binary classification problem, you might have one output neuron, while for a multi-class classification problem, you might have as many neurons as there are classes.

### 1.2.3. Hidden Layer
In order to obtain a prediction vector y, the network must perform certain mathematical operations. These operations are performed in the layers between the input and output layers which are called Hidden Layers.

These layers, aptly named because they are not directly observed from the input or output data, are pivotal in enabling deep learning models to represent and learn complex functions.

💡 **Fun Fact**  
The "deep" in deep learning or deep neural network usually refers to the multiple hidden layers found in the neural networks.

There are several types of hidden layers including: Dense (fully connected), convolutional layers, recurrent layers, normalization layers, etc.

1. **Dense (Fully Connected) Layers:**  
   These are the most common type where every neuron in the layer is connected to every neuron in the subsequent layer. The dense layer is a linear operation, but when combined with a non-linear activation function, it provides the network the capability to learn from the error and make adjustments, which is essential for learning complex patterns.

2. **Convolutional Layers:**  
   Primarily used in convolutional neural networks (CNNs), these layers are specifically designed to recognize spatial hierarchies in data, making them excellent for image recognition tasks. They work by sliding small windows (or kernels) over the input data to produce feature maps, capturing local patterns.

3. **Recurrent Layers:**  
   Found in recurrent neural networks (RNNs), these layers have loops that allow information to be passed from one step of the network to the next. This architecture is particularly beneficial for tasks where temporal dynamics and sequence recognition are crucial, like in natural language processing or time series prediction.

4. **Normalization Layers:**  
   These layers, such as Batch Normalization, adjust and scale the activations. By normalizing the layer's inputs, they can help maintain a stable mean and variance, improving the overall training speed and performance of deep networks by mitigating the problem of internal covariate shift.

## 1.2.3. Learning Process of a Neural Network:
A neural network learns by adjusting its weights based on the error in its predictions. The learning process consists of two main steps: **Forward Propagation** and **Backpropagation** ([Heskes T.M., 1991](https://journals.aps.org/pra/abstract/10.1103/PhysRevA.44.2718)).

### Forward Propagation:
1. **Input Layer:** The input layer receives the input data and passes it to the next layers.
2. **Hidden Layers:** The data is passed through one or more hidden layers, where it is processed using weights, biases, and activation functions.
3. **Output Layer:** The final layer outputs the network's prediction or result.

At this stage, the neural network makes a prediction based on the given input.

### Backpropagation (Geri Yayılım):
Backpropagation is the process by which the neural network learns from the error in its prediction.

1. **Error Calculation:** After forward propagation, the network calculates the error, which is the difference between the predicted value and the actual value (the ground truth).
2. **Gradient Calculation:** The network calculates the gradients (partial derivatives) of the error with respect to each weight in the network.
3. **Update Weights:** Using the gradients, the weights are adjusted to minimize the error. This is usually done using an optimization algorithm like **Gradient Descent**.

Backpropagation helps the network adjust the weights in the right direction to improve its predictions. This process is repeated for many iterations (epochs), allowing the network to gradually learn from the data and make more accurate predictions over time.

### Simple Summary:
- **Forward Propagation** is where the network makes predictions.
- **Backpropagation** is how the network learns by adjusting its weights based on the error in its predictions.

## 1.2.4. Loss Function:
Loss functions measure the difference between a neural network's predictions and the actual values. The goal of training is to minimize this loss for better accuracy ([Zhao H., 2015](https://arxiv.org/abs/1511.08861)).
![](https://miro.medium.com/v2/resize:fit:1400/0*QLXFJpiWv2qt2qcs.gif)

Image Source: [Medium](https://miro.medium.com/v2/resize:fit:1400/0*QLXFJpiWv2qt2qcs.gif)

### 1. **Mean Squared Error (MSE) Loss**
- **Definition:**  
  MSE calculates the average of the squared differences between the actual values ($y_i$) and predicted values ($\hat{y}_i$) ([Ma Y., 2019](https://www.spiedigitallibrary.org/conference-proceedings-of-spie/11209/112090I/Low-dose-CT-with-a-deep-convolutional-neural-network-blocks/10.1117/12.2542662.full)).

- **Formula:**  
  $$
  \text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
  $$

- **Usage:**  
  Used for regression tasks such as predicting house prices or temperatures.

- **Key Point:**  
  Squaring the errors amplifies large deviations, heavily penalizing significant errors.

![](https://miro.medium.com/v2/resize:fit:1400/1*n37x4cBieB5LBmOZCioSBw.gif)

Image Source: [Medium](https://miro.medium.com/v2/resize:fit:1400/1*n37x4cBieB5LBmOZCioSBw.gif)

---

### 2. **Cross Entropy Loss**
- **Definition:**  
  Measures the dissimilarity between predicted probabilities ($\hat{y}_i$) and true labels ($y_i$) ([Zhang Z., 2018](https://proceedings.neurips.cc/paper/2018/hash/f2925f97bc13ad2852a7a551802feea0-Abstract.html)).

- **Formula (Binary Classification):**  
  $$
  \text{Cross Entropy Loss} = -\frac{1}{n} \sum_{i=1}^n \Big(y_i \cdot \log(\hat{y}_i) + (1 - y_i) \cdot \log(1 - \hat{y}_i)\Big)
  $$

- **Usage:**  
  Standard for classification problems like image recognition or text analysis.

- **Key Point:**  
  Encourages the model to produce confident, correct predictions and penalizes incorrect overconfidence.

- **MSE:** Best for regression, minimizing the numerical error.

## 1.2.5. Gradient Descent

### What is Gradient Descent?
Gradient Descent is an optimization algorithm used to minimize the loss function of a machine learning model. It iteratively updates the model parameters by moving in the direction of the steepest descent (negative gradient) to find the minimum value of the loss function ([Velikanov M., 2021.](https://proceedings.neurips.cc/paper/2021/hash/14faf969228fc18fcd4fcf59437b0c97-Abstract.html))


---

### Gradient Descent Formula
The parameter $\theta$ is updated using the following rule ([Chizat L., 2020](https://proceedings.mlr.press/v125/chizat20a.html)):
$$
\theta = \theta - \alpha \frac{\partial L}{\partial \theta}
$$

Where:
- $\theta$: Model parameter (e.g., weights in neural networks)  
- $\alpha$: Learning rate (controls step size)  
- $\frac{\partial L}{\partial \theta}$: Gradient of the loss function $L$ with respect to $\theta$

---

### The Process
1. **Initialize Parameters:** Start with random values for $\theta$.  
2. **Compute Gradient:** Calculate $\frac{\partial L}{\partial \theta}$, the slope of the loss function with respect to $\theta$.  
3. **Update Parameters:** Adjust $\theta$ using the update rule above.  
4. **Repeat:** Continue updating until the loss converges to a minimum or meets a stopping criterion.

---

### Learning Rate ($\alpha$)
The learning rate determines the size of the steps taken towards the minimum ([Takase T., 2018](https://www.sciencedirect.com/science/article/pii/S0893608018300303)):
- **Small $\alpha$:** Slow convergence but more stable.  
- **Large $\alpha$:** Faster progress but may overshoot or diverge.  

Choosing an appropriate learning rate is crucial for the success of Gradient Descent.

---

### Variants of Gradient Descent
1. **Batch Gradient Descent:**  
   Computes the gradient using the entire dataset. It is stable but computationally expensive for large datasets. ([Andrychowicz M., 2016](https://proceedings.neurips.cc/paper/2016/hash/fb87582825f9d28a8d42c5e5e5e8b23d-Abstract.html))

2. **Stochastic Gradient Descent (SGD):**  
   Uses a single data point to update the parameters. It is faster but less stable, introducing noise into the optimization process. ([Amari S., 1993](https://www.sciencedirect.com/science/article/abs/pii/092523129390006O))

3. **Mini-batch Gradient Descent:**  
   Uses small batches of data points, balancing stability and speed. ([Hinton G., 2012]( http://www.cs.toronto.edu/~hinton/coursera/lecture6/lec6.pdf))
   
4. **Momentum:**
Accelerates gradient descent by considering the past gradients to smooth out updates, helping to navigate ravines and converge faster.([Qian N., 1999](https://www.sciencedirect.com/science/article/pii/S0893608098001166))

5. **Nesterov Accelerated Gradient (NAG):**
An improvement over Momentum, it computes the gradient at a lookahead position, leading to more informed updates. ([Su W., 2017](https://ieeexplore.ieee.org/abstract/document/8287318))

6. **Adagrad (Adaptive Gradient Algorithm):**
Adapts the learning rate for each parameter, performing larger updates for infrequent parameters and smaller updates for frequent ones. ([Gui Y., 2023](https://link.springer.com/article/10.1007/s10489-022-03629-7))

7. **RMSprop (Root Mean Square Propagation):**
Keeps a moving average of the squared gradients to adapt the learning rate, addressing Adagrad's diminishing learning rate issue. ([Kurbiel T., 2017](https://arxiv.org/abs/1708.01911))

8. **Adam (Adaptive Moment Estimation):**
Combines the ideas of Momentum and RMSprop, maintaining moving averages of both gradients and their squares for efficient optimization. ([Jais I., 2019](https://pdfs.semanticscholar.org/e4fd/4f1e6aef60ded196b22ae4ce9575e9506852.pdf))

## 1.2.6. Curse of Dimensionality

Before building a neural network, it is essential to understand a critical challenge in high-dimensional data analysis: the **Curse of Dimensionality**.

This term, coined by Richard Bellman in the 1960s, refers to the exponential increase in computational complexity and data sparsity as the number of dimensions (features) grows.

---

### Key Impacts of the Curse of Dimensionality ([Petrini L., 2023](https://arxiv.org/abs/2310.16154))

1. **Data Sparsity**  
   In high-dimensional spaces, data becomes sparse. To adequately sample all feature combinations, the required data grows exponentially.  
   - **Example**:  
     If 10 samples suffice for a single feature, you'd need:
       - 100 samples for 2 features  
       - 1,000 samples for 3 features  
       - And so on.  
   Without enough data, models struggle to generalize effectively.

2. **Distance Measures Lose Meaning**  
   Many machine learning algorithms rely on distance measures (e.g., Euclidean distance). However, in high dimensions:
   - Distances between points become nearly uniform.  
   - Clustering and nearest neighbor searches become less reliable.

3. **Overfitting**  
   Higher dimensions increase the model's flexibility, making it easier to fit noise rather than patterns in the data.  
   - **Result**: The model performs well on the training set but poorly on unseen data.

4. **Computational Complexity**  
   Algorithms that work efficiently in low dimensions can become computationally intractable in high-dimensional spaces.  
   - **Example**: Matrix operations (e.g., inversions) have a complexity that grows with dimensionality.

5. **Hughes Phenomenon**  
   Hughes (1968) observed that with a fixed number of training samples:
   - The predictive power of a classifier initially improves as dimensions increase.  
   - After a certain threshold, performance deteriorates due to data sparsity and overfitting.

---

### Addressing the Curse of Dimensionality ([Darbon J., 2020](https://link.springer.com/article/10.1007/s40687-020-00215-6))
1. **Feature Selection**  
   Identify and retain only the most relevant features for your task.  

2. **Dimensionality Reduction**  
   Techniques like **Principal Component Analysis (PCA)** or **t-SNE** can project high-dimensional data into lower-dimensional spaces while preserving critical patterns.

3. **Regularization**  
   Apply regularization techniques (e.g., L1, L2) to penalize model complexity and prevent overfitting.

4. **Increase Data Availability**  
   Collect more data to combat sparsity and improve model generalization.

## 1.3 Transfer Learning
**What is Transfer Learning?**

Transfer learning is a technique in machine learning where a model developed for a task is reused as the starting point for a model on a second task. This allows the second model to learn faster and achieve better performance because it can leverage the knowledge gained from the first task. ([Ribani R., 2019](https://ieeexplore.ieee.org/abstract/document/8920338))

**Why is it used?**

Transfer learning is used because it can save a lot of time and resources compared to training a model from scratch. Additionally, it can be used when there is limited data available for the second task, making it easier to train a good model. ([Pan S., 2014](https://www.taylorfrancis.com/chapters/edit/10.1201/b17320-26/transfer-learning-sinno-jialin-pan))

**When should we use it?**

Transfer learning should be used when the second task is related to the first task in some way. For example, if you have a model that has been trained to recognize objects in images and you want to train a model to recognize objects in videos, you can use the image recognition model as a starting point for the video recognition model. ([Weiss K., 2016](https://link.springer.com/article/10.1186/S40537-016-0043-6))

**Transfer Learning approaches**

There are several approaches to transfer learning, such as fine-tuning, feature extraction, and multi-task learning. Fine-tuning involves training a pre-trained model on the new task by adjusting the weights of the model. Feature extraction involves using the features learned by the pre-trained model and training a new classifier on top of it. Multi-task learning involves training a model on multiple tasks simultaneously, which can lead to better performance on both tasks. ([Lu J., 2015](https://www.sciencedirect.com/science/article/pii/S0950705115000179))

**Popular Architectures in Transfer Learning**

-   **VGG:**
    Known for its simplicity and deep architecture, VGG is widely used for image classification tasks and feature extraction.

-   **ResNet:**
    ResNet introduced skip connections to address the vanishing gradient problem, enabling the training of very deep networks with remarkable performance.

-  **MobileNet:**
    A lightweight architecture optimized for mobile and embedded devices, MobileNet is efficient and ideal for resource-constrained environments.

-   **EfficientNet**:
    Combines model scaling techniques to optimize performance and efficiency, achieving state-of-the-art results in various vision tasks with fewer parameters.

**Use Cases of Transfer Learning** ([Hosna A., 2022](https://link.springer.com/article/10.1186/s40537-022-00652-w))

-    **Image Classification:** Using pre-trained models like ResNet and EfficientNet to classify images in specialized datasets.
-    **Object Detection:** Leveraging architectures such as MobileNet for detecting and localizing objects in images.
-    **Natural Language Processing:** Adapting models like BERT or GPT for tasks such as text classification, sentiment analysis, or machine translation.
-   **Speech Recognition:** Applying transfer learning to models trained on large speech datasets to improve recognition in specific languages or accents.

## Model Structure and Layers Used

This deep learning model creates a convolutional neural network (CNN) on **TensorFlow** using the **Keras** library. The model is designed specifically for image classification problems. The layers and functions used are explained below.

### Layers and Parameters

1. **Input Layer**  
   - Defines the shape of the input data the model will receive.  
   - **shape=(224, 224, 3)**: Indicates that the input images are 224x224 pixels with 3 color channels (RGB).

2. **Conv2D (Convolutional Layer)**  
   - **Number of Filters (32)**: 32 filters are used in this layer.  
   - **Filter Size (3x3)**: Operations are performed on a 3x3 area of the image.  
   - **Activation Function**:  
     - `relu`: Helps learn nonlinear relationships by setting negative values to zero.  
     - Alternatives: `sigmoid` (for binary classification) or `tanh`.

3. **MaxPooling2D (Max Pooling Layer)**  
   - **Pooling Size (2x2)**: The maximum value from each section is selected, reducing the feature map.  
   - Reduces the computational load of the model and improves performance.

4. **Flatten (Flattening Layer)**  
   - Converts 2D image data into a 1D vector.  
   - This step is necessary for working with fully connected layers.

5. **Dense (Fully Connected Layer)**  
   - **Number of Neurons (128)**: Enables the learning of more complex features.  
   - **Activation Function**: `relu`.

6. **Dropout (Dropout Layer)**  
   - **Rate (0.5)**: Disables random neurons during each training step to prevent overfitting.

7. **Output Layer**  
   - **Number of Neurons (2)**: Indicates that the model will classify two classes.  
   - **Activation Function**:  
     - `softmax`: Produces probabilities for each class and is used for multi-class classification.  
     - Alternative: `sigmoid` (for binary classification).

### Model Compilation
- **Optimization Algorithm**:  
  - `adam`: A fast optimization algorithm that automatically adjusts the learning rate.  
  - Alternative: `sgd` (Stochastic Gradient Descent).  
- **Loss Function**:  
  - `sparse_categorical_crossentropy`: For multi-class classification problems.  
  - Alternative: `categorical_crossentropy` (when labels are one-hot encoded).  
- **Metric**: `accuracy`.

### Model Summary
The model structure, connections between layers, and information like the number of parameters can be summarized using `model.summary()`.


````python
# Import necessary modules from TensorFlow Keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input

# Define a sequential model with layers
model = Sequential([
    # Input layer with shape (224, 224, 3) for RGB images
    Input(shape=(224, 224, 3)),

    # First convolutional layer with 32 filters, 3x3 kernel, and ReLU activation
    Conv2D(32, (3, 3), activation='relu'),
    # First max pooling layer with 2x2 pool size
    MaxPooling2D((2, 2)),

    # Second convolutional layer with 64 filters, 3x3 kernel, and ReLU activation
    Conv2D(64, (3, 3), activation='relu'),
    # Second max pooling layer with 2x2 pool size
    MaxPooling2D((2, 2)),

    # Third convolutional layer with 128 filters, 3x3 kernel, and ReLU activation
    Conv2D(128, (3, 3), activation='relu'),
    # Third max pooling layer with 2x2 pool size
    MaxPooling2D((2, 2)),

    # Flatten the feature map into a vector
    Flatten(),

    # Fully connected layer with 128 neurons and ReLU activation
    Dense(128, activation='relu'),
    # Dropout layer with a rate of 0.5 to prevent overfitting
    Dropout(0.5),

    # Output layer with 2 neurons and softmax activation for binary classification
    Dense(2, activation='softmax')
])

# Compile the model with Adam optimizer, sparse categorical crossentropy loss, and accuracy metric
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()
````

# ImageDataGenerator and Data Preparation

This code block is written for preprocessing data and creating training/validation datasets for training an image classification model. The `ImageDataGenerator` class from TensorFlow/Keras is used.

### 1. **Defining ImageDataGenerator**
- **`rescale=1./255`**: Normalizes the pixel values of the image from the range 0-255 to 0-1. This helps the model learn more effectively.
- **`validation_split=0.2`**: Splits the dataset so that 20% is used for validation and 80% is used for training.

### 2. **Loading Training Data**
- **`train_path`**: The directory where the training data is located.
- **`target_size=(224, 224)`**: Resizes the images to 224x224 pixels.
- **`batch_size=32`**: The data is grouped in batches of 32 images per step.
- **`class_mode='binary'`**: Specifies that binary classification will be performed.
- **`subset='training'`**: This subset refers to the training data.

### 3. **Loading Validation Data**
- The parameters above are similar to those for the training data, but **`subset='validation'`** is used to load the validation data.

### Summary
This code loads the training and validation data from a directory, normalizes the images, and rescales them to the specified size. The data is dynamically loaded using the `flow_from_directory` method and is appropriately split. This prepares the data for model training and validation.


````python
# Import the ImageDataGenerator class for image augmentation and preprocessing
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create an ImageDataGenerator for training with rescaling and validation split
train_gen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

# Generate training data with images rescaled, resized, and partitioned for training
train_data = train_gen.flow_from_directory(
    train_path,                # Path to the training dataset
    target_size=(224, 224),    # Resize images to 224x224 pixels
    batch_size=32,             # Number of images in each batch
    class_mode='binary',       # Binary classification mode
    subset='training'          # Use the training subset
)

# Generate validation data with the same preprocessing
val_data = train_gen.flow_from_directory(
    train_path,                # Path to the training dataset
    target_size=(224, 224),    # Resize images to 224x224 pixels
    batch_size=32,             # Number of images in each batch
    class_mode='binary',       # Binary classification mode
    subset='validation'        # Use the validation subset
)

````

# Creating a MobileNetV2-Based Deep Learning Model

This code creates an image classification model using the **MobileNetV2** architecture. The base model (pre-trained model) is used with transfer learning, and custom fully connected layers are added.

### 1. **MobileNetV2 Base Model**
```python
# Import the MobileNetV2 model pre-trained on ImageNet
from tensorflow.keras.applications import MobileNetV2

# Load the MobileNetV2 model with pre-trained weights, excluding the top classification layers
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the base model to prevent its weights from being updated during training
base_model.trainable = False

```



````python
# Import necessary modules for building and modifying a model
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout

# Load the MobileNetV2 model pre-trained on ImageNet, excluding the top classification layers
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the base model to prevent its weights from being updated during training
base_model.trainable = False

# Add custom layers on top of the base model
x = base_model.output  # Output from the base model
x = GlobalAveragePooling2D()(x)  # Add a global average pooling layer
x = Dense(128, activation='relu')(x)  # Add a dense layer with 128 neurons and ReLU activation
x = Dropout(0.5)(x)  # Add a dropout layer with a rate of 0.5 to prevent overfitting
predictions = Dense(1, activation='sigmoid')(x)  # Add a final dense layer with a sigmoid activation for binary classification

# Create the final model by specifying the input and output layers
model = Model(inputs=base_model.input, outputs=predictions)

# Compile the model with Adam optimizer, binary cross-entropy loss, and accuracy metric
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
````

````python
# Import the EarlyStopping callback from TensorFlow Keras
from tensorflow.keras.callbacks import EarlyStopping

# Define the EarlyStopping callback to stop training when validation loss stops improving
early_stopping = EarlyStopping(
    monitor='val_loss',          # Monitor the validation loss
    patience=3,                  # Number of epochs with no improvement before stopping
    restore_best_weights=True    # Restore the weights of the best epoch
)

# Train the model with training and validation data, applying early stopping
history = model.fit(
    train_data,                  # Training data
    validation_data=val_data,    # Validation data
    epochs=10,                   # Maximum number of epochs
    batch_size=32,               # Batch size for training
    callbacks=[early_stopping]   # List of callbacks to use during training
)
````

````python
# Create an ImageDataGenerator for preprocessing the test data by rescaling pixel values
test_gen = ImageDataGenerator(rescale=1./255)

# Load the test data with preprocessing, resizing, and batching
test_data = test_gen.flow_from_directory(
    test_path,                # Path to the test dataset
    target_size=(224, 224),   # Resize images to 224x224 pixels
    batch_size=32,            # Number of images in each batch
    class_mode='binary'       # Binary classification mode
)

# Evaluate the model on the test dataset
test_loss, test_acc = model.evaluate(test_data)

# Print the test accuracy
print(f"Test Accuracy: {test_acc:.2f}")
````

````python
# Import necessary libraries and modules
import matplotlib.pyplot as plt
from tensorflow.keras.applications import ResNet50, VGG16, InceptionV3, EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping

# Set up an image data generator for training and validation datasets
train_gen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

# Create training data generator
train_data = train_gen.flow_from_directory(
    train_path,                # Path to training dataset
    target_size=(224, 224),    # Resize images to 224x224 pixels
    batch_size=32,             # Batch size for training
    class_mode='binary',       # Binary classification mode
    subset='training'          # Training subset
)

# Create validation data generator
val_data = train_gen.flow_from_directory(
    train_path,                # Path to training dataset
    target_size=(224, 224),    # Resize images to 224x224 pixels
    batch_size=32,             # Batch size for validation
    class_mode='binary',       # Binary classification mode
    subset='validation'        # Validation subset
)

# Define a function to create a model by adding custom layers on top of a base model
def create_model(base_model):
    x = base_model.output
    x = GlobalAveragePooling2D()(x)  # Add global average pooling
    x = Dense(128, activation='relu')(x)  # Add a dense layer with ReLU activation
    x = Dropout(0.5)(x)  # Add a dropout layer to prevent overfitting
    predictions = Dense(1, activation='sigmoid')(x)  # Output layer with sigmoid activation
    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])  # Compile the model
    return model

# Initialize pre-trained base models with weights and without the top classification layers
models = {
    'ResNet50': ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3)),
    'VGG16': VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3)),
    'InceptionV3': InceptionV3(weights='imagenet', include_top=False, input_shape=(224, 224, 3)),
    'EfficientNetB0': EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3)),
}

# Dictionary to store the training history of each model
history_dict = {}

# Set up early stopping to stop training if validation loss doesn't improve
early_stopping = EarlyStopping(
    monitor='val_loss',          # Monitor validation loss
    patience=3,                  # Number of epochs without improvement to stop training
    restore_best_weights=True    # Restore the weights of the best epoch
)

# Train each model and store its training history
for name, base_model in models.items():
    base_model.trainable = False  # Freeze the base model layers
    model = create_model(base_model)  # Create a new model using the base model

    print(f"Starting Training: {name}")
    history = model.fit(
        train_data,              # Training data
        validation_data=val_data, # Validation data
        epochs=10,               # Number of epochs
        callbacks=[early_stopping]  # Early stopping callback
    )

    history_dict[name] = history  # Store the history of the model

# Plot training and validation accuracy for all models
plt.figure(figsize=(14, 6))

# Subplot for accuracy
plt.subplot(1, 2, 1)
for name, history in history_dict.items():
    plt.plot(history.history['accuracy'], label=f'{name} - Training Accuracy')
    plt.plot(history.history['val_accuracy'], label=f'{name} - Validation Accuracy')
plt.title('Model Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Subplot for loss
plt.subplot(1, 2, 2)
for name, history in history_dict.items():
    plt.plot(history.history['loss'], label=f'{name} - Training Loss')
    plt.plot(history.history['val_loss'], label=f'{name} - Validation Loss')
plt.title('Model Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()

````

**(5) Add Early Stopping**

````python
class EarlyStopping:
    def __init__(self, patience=3, min_delta=0.001):
        self.patience = patience
        self.min_delta = min_delta
        self.best_loss = float('inf')
        self.counter = 0

    def should_stop(self, loss):
        if loss < self.best_loss - self.min_delta:
            self.best_loss = loss
            self.counter = 0
        else:
            self.counter += 1

        return self.counter >= self.patience
````

**(6) Train the Model with Early Stopping**

````python
num_epochs = 20  # Set higher epochs, but early stopping will stop it automatically
early_stopping = EarlyStopping(patience=5)

train_losses = []
test_losses = []

for epoch in range(num_epochs):
    model.train()
    total_loss = 0

    for features, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(features)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    train_losses.append(avg_loss)

    # Validate Model
    model.eval()
    test_loss = 0
    with torch.no_grad():
        for features, labels in test_loader:
            outputs = model(features)
            loss = criterion(outputs, labels)
            test_loss += loss.item()

    avg_test_loss = test_loss / len(test_loader)
    test_losses.append(avg_test_loss)

    print(f"Epoch [{epoch+1}/{num_epochs}] - Train Loss: {avg_loss:.4f} | Test Loss: {avg_test_loss:.4f}")

    if early_stopping.should_stop(avg_test_loss):
        print("⏹️ Early Stopping Triggered!")
        break
````

# 3. Introduction to Generative Adversarial Networks (GANs)
**What is a Generative Adversarial Network (GAN)?**

A GAN is a class of machine learning models that consists of two neural networks working in opposition to each other ([Wang K., 2017](https://ieeexplore.ieee.org/abstract/document/8039016)):

-    **Generator:** Creates new data samples similar to the training data.
-    **Discriminator:** Evaluates whether a sample is real (from the training data) or fake (created by the Generator).

**Types of GANs** ([Wu A., 2022](https://www.sciencedirect.com/science/article/pii/S0360132322007089))

-   **Vanilla GAN:**
    This is the most basic form of GAN, where both the Generator and Discriminator are simple multi-layer perceptrons. The primary objective of the Vanilla GAN is to optimize its mathematical equation using stochastic gradient descent. It serves as the foundational framework for more advanced GAN variants.

-    **Conditional GAN (CGAN):**
    In CGAN, additional conditional parameters are incorporated into the model to guide the data generation process.
      -  A parameter ‘y’ is added to the Generator to create data corresponding to specific conditions.
      -  The same conditional information is provided to the Discriminator, helping it differentiate between real and fake data more effectively.

-   **Deep Convolutional GAN (DCGAN):**
    DCGAN is one of the most widely used and successful GAN architectures. It replaces the multi-layer perceptrons with Convolutional Neural Networks (ConvNets).
      -  Instead of max pooling, it uses convolutional strides.
      -  The network avoids fully connected layers, relying on ConvNets for feature extraction.
      -  DCGAN is particularly effective for tasks like image generation and feature learning.

-   **Laplacian Pyramid GAN (LAPGAN):**
    LAPGAN leverages a Laplacian pyramid, which represents images through a set of band-pass filters and a low-frequency residual.
      -  Multiple Generators and Discriminators operate at various levels of the pyramid.
      -  During the forward pass, images are down-sampled layer by layer, and in the backward pass, they are up-scaled, gaining additional details with the help of Conditional GANs.
      -  This technique is particularly effective for producing high-quality images.

-    **Super Resolution GAN (SRGAN):**
    SRGAN focuses on generating high-resolution images from low-resolution inputs.
      -  It combines a deep neural network with an adversarial network to improve image details.
      - SRGAN is especially useful for tasks requiring enhanced image resolution, ensuring details are preserved while minimizing artifacts or errors.

The two networks engage in a game-like scenario where the Generator tries to fool the Discriminator, and the Discriminator tries to identify the fake samples.

## 3.1 Architecture of GANs
The generator in a GAN is tasked with creating synthetic data, such as images or text, that mimics real data. It takes random noise as input and transforms it into a realistic data sample through layers of learnable parameters. By capturing the underlying distribution of the training data, the generator refines its parameters using backpropagation during training. The ultimate goal of the generator is to produce diverse and high-quality samples that convincingly resemble real data, effectively fooling the discriminator. ([Wang Z., 2021](https://dl.acm.org/doi/abs/10.1145/3439723))

**Generator Model**

The generator is responsible for creating new data, such as images or text, that closely resembles real data.

  - Input: Random noise.
  -  Output: A synthetic data sample.
  -  How it works: Through layers of learnable parameters, the generator captures the underlying distribution of the training data. It uses backpropagation to refine its parameters, improving the realism of the generated samples.
  -  Goal: Produce diverse, high-quality samples that can successfully fool the discriminator.

**Generator Loss:**
The generator's loss function is designed to maximize the likelihood of the discriminator being fooled. It is defined as ([Veiner J., 2024](https://www.mdpi.com/1099-4300/26/4/290)):  

$$
J_G = -\frac{1}{m} \sum_{i=1}^m \log D(G(z_i))
$$

Where:  
\( J_G \): Measures how effectively the generator fools the discriminator.  
\( D(G(z_i)) \): Represents the discriminator's confidence that the generated sample is real.  

**Discriminator Model**

The discriminator acts as a binary classifier that distinguishes real data from fake data generated by the generator.

   - Input: Real samples from the dataset and fake samples from the generator.
   - Output: A probability indicating whether a sample is real or fake.
   - How it works: Over time, the discriminator learns to identify the subtle differences between real and generated data, refining its parameters through training.
   - Goal: Maximize accuracy in distinguishing real samples from fake ones.

**Discriminator Loss:**
The discriminator minimizes its loss function, which evaluates its ability to correctly classify both real and fake samples. The loss function is ([Zadorozhnyy V., 2021](https://openaccess.thecvf.com/content/CVPR2021/html/Zadorozhnyy_Adaptive_Weighted_Discriminator_for_Training_Generative_Adversarial_Networks_CVPR_2021_paper.html)):
$$
JD=−1m∑i=1mlog⁡D(xi)−1m∑i=1mlog⁡(1−D(G(zi)))
$$

Where:

(J_D): Measures the discriminator’s ability to differentiate real and fake samples.

(log D(x_i)): Likelihood of correctly classifying real data.

(log (1 - D(G(z_i)))): Likelihood of correctly identifying fake data.

The discriminator strives to minimize this loss by accurately distinguishing between real and generated samples.

**Minimax Loss**

The adversarial training process in GANs can be described as a minimax game, where:

$$
min⁡Gmax⁡DV(G,D)=Ex∼pdata(x)[log⁡D(x)]+Ez∼pz(z)[log⁡(1−D(G(z)))]
$$

Where:

 -   G: Generator network.
 -   D: Discriminator network.
 -   pdata(x): True data distribution.
 -   pz(z): Noise distribution.
 -   D(x): Probability that the discriminator classifies real data correctly.
 -   D(G(z)): Probability that the discriminator identifies generated data as real.

The generator minimizes this loss, while the discriminator maximizes it, resulting in a dynamic adversarial process that improves both networks over time.

**Why are GANs Important?**

-    GANs can generate realistic images, videos, and audio.
-    They are used in applications like image generation, style transfer, data augmentation, and more.
-    GANs provide insights into how machines can mimic human creativity.

**How Does a GAN Work?**

The process of how a GAN operates involves the following steps:

-    **Initialization:**
      -  Generator (G): Responsible for creating new data, such as images or text, that mimics real data.
      -  Discriminator (D): Acts as a judge, distinguishing between real data (from the training dataset) and fake data generated by G.

-    **Generator's First Step:**
    The Generator takes a random noise vector as input.
       - This noise serves as the foundation for G to create a new data sample.
       - G uses its internal network layers and learned patterns to transform the noise into data resembling real samples, such as an image.

-    **Discriminator's Role:**
    The Discriminator evaluates two types of inputs:
      -  Real data from the training dataset.
      -  Fake data created by G.
      -  It analyzes each sample and assigns a probability score between 0 and 1, where 1 indicates the data is real, and 0 suggests it’s fake.

-   **Adversarial Learning Process:**
    - If D accurately classifies real data as real (close to 1) and fake data as fake (close to 0), it performs well, but improvement is gradual.
    -  The competition begins as G tries to produce fake data realistic enough to trick D, while D aims to sharpen its ability to differentiate between real and fake data.

-    **Generator's Learning:**
    -    If D mistakenly identifies G’s fake data as real (score close to 1), it signals that G is improving.
    -    G receives positive feedback to further enhance its ability to create realistic data.
     -   Conversely, D is penalized for being fooled, encouraging it to refine its detection skills.

-    **Discriminator's Adaptation:**
     -   If D correctly identifies G’s fake data as fake (score close to 0), it is rewarded for its accuracy.
     -   G, however, gets no reward, prompting it to work harder to create more convincing fake data.

-    **Iterative Refinement:**
      -  This constant back-and-forth competition between G and D helps both networks improve over time.
      -  G learns to produce highly realistic data, and D becomes more adept at distinguishing real from fake samples.

````python
# Load the CIFAR-10 training dataset with the defined transformations
train_dataset = datasets.CIFAR10(
    root='./data',           # Directory to store the dataset
    train=True,              # Load the training set
    download=True,           # Download the dataset if not already available
    transform=transform      # Apply the defined transformations
)

# Create a DataLoader to manage batches and shuffle the training data
dataloader = torch.utils.data.DataLoader(
    train_dataset,           # Dataset to load
    batch_size=32,           # Number of samples per batch
    shuffle=True             # Shuffle the dataset at every epoch
)
````

## Step 4: Defining parameters to be used in later processes

````python
# Hyperparameters for the model
latent_dim = 100  # Dimensionality of the latent space (input noise vector for GANs)
lr = 0.0002       # Learning rate for the optimizer
beta1 = 0.5       # Beta1 hyperparameter for the Adam optimizer (momentum decay)
beta2 = 0.999     # Beta2 hyperparameter for the Adam optimizer (squared gradient decay)
num_epochs = 10   # Number of training epochs
````

## Step 5: Defining a Utility Class to Build the Generator

````python
# Define the Generator class
class Generator(nn.Module):
    def __init__(self, latent_dim):
        super(Generator, self).__init__()

        # Define the generator model as a sequential block
        self.model = nn.Sequential(
            nn.Linear(latent_dim, 128 * 8 * 8),  # Fully connected layer to project latent space
            nn.ReLU(),                           # Activation function
            nn.Unflatten(1, (128, 8, 8)),        # Reshape the output to 128 channels of 8x8
            nn.Upsample(scale_factor=2),         # Upsample the feature map by a factor of 2
            nn.Conv2d(128, 128, kernel_size=3, padding=1),  # Convolutional layer
            nn.BatchNorm2d(128, momentum=0.78),  # Batch normalization
            nn.ReLU(),                           # Activation function
            nn.Upsample(scale_factor=2),         # Further upsample the feature map
            nn.Conv2d(128, 64, kernel_size=3, padding=1),  # Convolutional layer
            nn.BatchNorm2d(64, momentum=0.78),   # Batch normalization
            nn.ReLU(),                           # Activation function
            nn.Conv2d(64, 3, kernel_size=3, padding=1),  # Final convolutional layer to output image
            nn.Tanh()                            # Tanh activation to scale pixel values to [-1, 1]
        )

    # Define the forward pass for the generator
    def forward(self, z):
        img = self.model(z)  # Generate the image from the input latent vector
        return img
````

## Step 6: Defining a Utility Class to Build the Discriminator

````python
# Define the Discriminator class
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()

        # Define the discriminator model as a sequential block
        self.model = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, stride=2, padding=1),  # Convolutional layer for feature extraction
            nn.LeakyReLU(0.2),                                    # LeakyReLU activation for better gradient flow
            nn.Dropout(0.25),                                     # Dropout to prevent overfitting
            nn.Conv2d(32, 64, kernel_size=3, stride=2, padding=1),# Convolutional layer
            nn.ZeroPad2d((0, 1, 0, 1)),                           # Zero padding for spatial dimensions
            nn.BatchNorm2d(64, momentum=0.82),                    # Batch normalization for stable training
            nn.LeakyReLU(0.25),                                   # LeakyReLU activation
            nn.Dropout(0.25),                                     # Dropout layer
            nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1),# Convolutional layer
            nn.BatchNorm2d(128, momentum=0.82),                   # Batch normalization
            nn.LeakyReLU(0.2),                                    # LeakyReLU activation
            nn.Dropout(0.25),                                     # Dropout layer
            nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1),# Convolutional layer
            nn.BatchNorm2d(256, momentum=0.8),                    # Batch normalization
            nn.LeakyReLU(0.25),                                   # LeakyReLU activation
            nn.Dropout(0.25),                                     # Dropout layer
            nn.Flatten(),                                         # Flatten the feature map to a vector
            nn.Linear(256 * 5 * 5, 1),                            # Fully connected layer for binary classification
            nn.Sigmoid()                                          # Sigmoid activation for outputting a probability
        )

    # Define the forward pass for the discriminator
    def forward(self, img):
        validity = self.model(img)  # Predict the validity of the input image
        return validity
````

## Step 7: Building the Generative Adversarial Network

````python
# Define the generator and discriminator
# Initialize the generator and move it to the appropriate device (GPU/CPU)
generator = Generator(latent_dim).to(device)

# Initialize the discriminator and move it to the appropriate device (GPU/CPU)
discriminator = Discriminator().to(device)

# Define the adversarial loss function (Binary Cross-Entropy Loss)
adversarial_loss = nn.BCELoss()

# Define the optimizer for the generator
optimizer_G = optim.Adam(
    generator.parameters(),    # Parameters to optimize
    lr=lr,                     # Learning rate
    betas=(beta1, beta2)       # Beta coefficients for Adam optimizer
)

# Define the optimizer for the discriminator
optimizer_D = optim.Adam(
    discriminator.parameters(), # Parameters to optimize
    lr=lr,                      # Learning rate
    betas=(beta1, beta2)        # Beta coefficients for Adam optimizer
)
````

## Step 8: Training the Generative Adversarial Network

````python
# Training loop
for epoch in range(num_epochs):
    for i, batch in enumerate(dataloader):
        # Extract the real images from the batch and move to the device
        real_images = batch[0].to(device)

        # Create ground truths for adversarial loss
        valid = torch.ones(real_images.size(0), 1, device=device)  # Label for real images
        fake = torch.zeros(real_images.size(0), 1, device=device)  # Label for fake images

        # ---------------------
        #  Train Discriminator
        # ---------------------
        optimizer_D.zero_grad()  # Zero the gradient of the discriminator optimizer

        # Sample noise as generator input
        z = torch.randn(real_images.size(0), latent_dim, device=device)

        # Generate a batch of fake images
        fake_images = generator(z)

        # Calculate discriminator loss on real images
        real_loss = adversarial_loss(discriminator(real_images), valid)

        # Calculate discriminator loss on fake images
        fake_loss = adversarial_loss(discriminator(fake_images.detach()), fake)

        # Average the real and fake losses
        d_loss = (real_loss + fake_loss) / 2

        # Backward pass and optimize discriminator
        d_loss.backward()
        optimizer_D.step()

        # -----------------
        #  Train Generator
        # -----------------
        optimizer_G.zero_grad()  # Zero the gradient of the generator optimizer

        # Generate a batch of fake images
        gen_images = generator(z)

        # Calculate adversarial loss for the generator
        g_loss = adversarial_loss(discriminator(gen_images), valid)

        # Backward pass and optimize generator
        g_loss.backward()
        optimizer_G.step()

        # ---------------------
        #  Progress Monitoring
        # ---------------------
        if (i + 1) % 100 == 0:  # Log progress every 100 batches
            print(
                f"Epoch [{epoch+1}/{num_epochs}] "
                f"Batch {i+1}/{len(dataloader)} "
                f"Discriminator Loss: {d_loss.item():.4f} "
                f"Generator Loss: {g_loss.item():.4f}"
            )

    # Save and display generated images every 10 epochs
    if (epoch + 1) % 10 == 0:
        with torch.no_grad():  # Disable gradient computation for inference
            z = torch.randn(16, latent_dim, device=device)  # Generate noise
            generated = generator(z).detach().cpu()        # Generate images and move to CPU
            grid = torchvision.utils.make_grid(generated, nrow=4, normalize=True)  # Create grid of images
            plt.imshow(np.transpose(grid, (1, 2, 0)))      # Transpose dimensions for visualization
            plt.axis("off")                                # Turn off axis
            plt.show()                                     # Display generated images
````
